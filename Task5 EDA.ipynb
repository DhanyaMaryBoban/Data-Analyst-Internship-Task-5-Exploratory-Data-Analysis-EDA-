4. How do gross sales, units sold and sale price interact?

```python
sns.pairplot(df[['gross sales', 'units sold', 'sale price']])
plt.show()
```

5. What are the strongest correlations among numerical features? (eg: unit sold, gross sales, publisher revenue and sales rank)?

```python
corr = df[['units sold', 'gross sales','publisher revenue','sales rank']].corr()
```

```python
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

#### Observation:
- Here unit sold and sales rank are positively correlated.
- Similarily, Gross sales and publisher revenue are positively correlated.

6. What is the distribution of units sold?

```python
sns.histplot(df['units sold'], bins = 20, kde = True)
plt.title('Distribution of Units Sold')
plt.show()
```

#### Observation:
- Over 350 books sold between 0 and 10,000 copies.This means the majority of books in the dataset don’t sell huge volumes.
- Fewer than 50 books sold between 30,000 and 60,000 copies. These are the rare bestsellers that stand out from the rest.
- The chart is right-skewed. In simple terms, most books sell modestly, while only a handful sell extremely well.

7. Which genres have the highest variablility in gross sales?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='genre', y='gross sales', data=df)
plt.title("Gross Sales Distribution by Genre")
plt.xlabel("Genre")
plt.ylabel("Gross Sales")
plt.xticks(rotation=45)  
plt.show()
```

#### Observation
- Genre fiction earn much higher or much lower sales than the rest.
- Nonfiction & Fiction earn similar sales, but there are still some big sellers or very low sellers.
- Children’s books  are very consistent — no extreme highs or lows.

8. Is there a trend between book average rating and gross sales?

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Book_average_rating', y='gross sales', data=df)

plt.title("Book Average Rating vs Gross Sales")
plt.xlabel("Book Average Rating")
plt.ylabel("Gross Sales")
plt.show()
```

#### Observation:
- Most books have ratings between 3.0 and 4.75, but their gross sales remain below 10,000.
- Only a few books with ratings between 3.25 and 4.0 achieve gross sales between 20,000 and 50,000.
- This shows that higher ratings do not automatically lead to higher sales.
